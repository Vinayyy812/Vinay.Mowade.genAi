(Q3) **Insurance Company Model **
You need to create a model to hold information about the insurer's vehicle and calculate the premiums (monthly, quarterly, yearly) based on the number of years of insurance, with the vehicle's value depreciating by 7% annually.

**Steps:**
1) Define a class for the vehicle insurance policy.

2) Add attributes for vehicle value, number of insurance years, and depreciation rate (7% per year).

3) Calculate the vehicle value after depreciation for each year.

4) Generate monthly, quarterly, and yearly premium amounts based on the depreciated value.

5) Output the premium values in a chart.

```python
import matplotlib.pyplot as plt

class VehicleInsurance:
    def __init__(self, initial_value, years):
        self.initial_value = initial_value
        self.years = years
        self.depreciation_rate = 0.07  # 7% annual depreciation
        self.depreciated_values = self.calculate_depreciation()

    def calculate_depreciation(self):
        values = []
        current_value = self.initial_value
        for year in range(self.years):
            current_value -= current_value * self.depreciation_rate
            values.append(current_value)
        return values

    def calculate_premiums(self):
        yearly_premiums = [value * 0.05 for value in self.depreciated_values]  # 5% premium rate
        quarterly_premiums = [premium / 4 for premium in yearly_premiums]
        monthly_premiums = [premium / 12 for premium in yearly_premiums]
        return yearly_premiums, quarterly_premiums, monthly_premiums

    def display_premiums(self):
        years = list(range(1, self.years + 1))
        yearly_premiums, quarterly_premiums, monthly_premiums = self.calculate_premiums()

        plt.plot(years, yearly_premiums, label='Yearly Premiums')
        plt.plot(years, quarterly_premiums, label='Quarterly Premiums')
        plt.plot(years, monthly_premiums, label='Monthly Premiums')

        plt.xlabel('Years of Insurance')
        plt.ylabel('Premium Amount')
        plt.title('Vehicle Insurance Premiums over Years')
        plt.legend()
        plt.show()

# Example usage:
vehicle = VehicleInsurance(initial_value=100000, years=5)
vehicle.display_premiums()
```
